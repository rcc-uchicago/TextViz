```python
Ham_text = open("../corpus/Ham_t.txt").read()
```

```python
import spacy
nlp = spacy.load('en_core_web_sm')
```

```python
Ham_spacy = nlp(Ham_text)
```

```python
for token in Ham_spacy:
    print(token.text, token.pos_, token.tag_, token.lemma_) 
#          token.dep_, token.shape_, token.is_alpha, token.is_stop)
```

```python
Ham_spacy_tokens = list(item.text for item in Ham_spacy)
Ham_spacy_pos = list(item.pos_ for item in Ham_spacy)
Ham_spacy_tags = list(item.tag_ for item in Ham_spacy)
Ham_spacy_lemmas = list(item.lemma_ for item in Ham_spacy)
```

```python
import pandas
Ham_spacy_pandasdf = pandas.DataFrame({'Tokens':Ham_spacy_tokens, 'POS':Ham_spacy_pos, 'POS Tag':Ham_spacy_tags,
                                       'Lemmas':Ham_spacy_lemmas})                                       
```
