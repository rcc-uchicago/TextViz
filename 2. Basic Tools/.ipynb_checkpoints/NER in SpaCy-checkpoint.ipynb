```python
import spacy
from spacy import displacy
from collections import Counter
from pprint import pprint
```

```python
Ham_text = open("../corpus/Ham_t.txt").read()
```

```python
# SpaCy NER

nlp = spacy.load('en_core_web_sm')
Ham_doc = nlp(Ham_text)

for ent in Ham_doc.ents:
    print(ent.text, ent.start_char, ent.end_char, ent.label_)
```

```python
spacy_Ham_labels = [x.label_ for x in Ham_doc.ents]
pprint(Counter(spacy_Ham_labels))
```

```python
pprint([(X.text, X.label_) for X in Ham_doc.ents])
```

```python
pprint([(X, X.ent_iob_, X.ent_type_) for X in Ham_doc if X.ent_type_ != ''])
```

```python
# sentence example
sentences = [x for x in Ham_doc.sents]
print(sentences[24])
```

```python
displacy.render(nlp(str(sentences[24])), jupyter=True, style='ent')
```

```python
displacy.render(nlp(str(sentences[24])), style='dep', jupyter = True, options = {'distance': 120})
```
